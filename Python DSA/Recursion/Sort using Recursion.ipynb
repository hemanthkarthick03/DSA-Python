# Sorting using Recursion

```python
def insert(lst, temp):
    if not lst or lst[-1] <= temp:
        lst.append(temp)
        return
    
    val = lst.pop()
    insert(lst, temp)
    
    lst.append(val)
```

```python
def sort_list(lst):
    if lst:
        temp = lst.pop()
        sort_list(lst)
        insert(lst, temp)
```

```python
lst = [34, 3, 31, 98, 92, 23]
sort_list(lst)

lst
```
