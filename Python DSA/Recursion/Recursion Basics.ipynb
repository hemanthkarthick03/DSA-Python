## Recursion

```python
def print_n_to_1(n):
    if n == 0:
        return

    print(n)
    print_n_to_1(n - 1)

print_n_to_1(5)
```

```python
def print_1_to_n(n):
    if n == 0:
        return

    print_1_to_n(n - 1)
    print(n)

print_1_to_n(5)
```

```python
def print_1_to_n_add(i, n):
    if i > n:
        return 
    
    print_1_to_n_add(i+1, n)
    print(i)

print_1_to_n_add(1, 5)
```

```python
def print_n_to_1_add(i, n):
    if i > n:
        return
    
    print(i)
    print_n_to_1_add(i + 1, n)

print_n_to_1_add(1, 5)
```

### Good Number

```python
from math import ceil, floor

def countGoodNumbers(n: int) -> int:
    even_position = ceil(n/2)
    odd_position = floor(n/2)

    return 5 ** even_position * 4 ** odd_position % (10**9 + 7)

countGoodNumbers(4)
```
