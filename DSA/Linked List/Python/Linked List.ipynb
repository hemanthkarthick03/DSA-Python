# Linked List

```python
class ListNode:
    def __init__(self, data):
        self.data = data
        self.next = None
```

## Insertion

- At Beginning
- At Middle
- At End

### At Beginning

```python
head = None

def insert_at_beginning(data):
    new_node = ListNode(data)
    head = new_node
```

### At End

```python
def insert_at_end(data):
    new_node = ListNode(data)

    if head is None:
        head = new_node
        return

    current = head
    
    while current.next:
        current = current.next

    current.next = new_node
```
