# [1790. Check if One String Swap Can Make Strings Equal](https://leetcode.com/problems/check-if-one-string-swap-can-make-strings-equal/description/)

You are given two strings <code>s1</code> and <code>s2</code> of equal length. A **string swap**  is an operation where you choose two indices in a string (not necessarily different) and swap the characters at these indices.

Return <code>true</code> if it is possible to make both strings equal by performing **at most one string swap ** on **exactly one**  of the strings. Otherwise, return <code>false</code>.

**Example 1:** 

```
Input: s1 = "bank", s2 = "kanb"
Output: true
Explanation: For example, swap the first character with the last character of s2 to make "bank".
```

**Example 2:** 

```
Input: s1 = "attack", s2 = "defend"
Output: false
Explanation: It is impossible to make them equal with one string swap.
```

**Example 3:** 

```
Input: s1 = "kelb", s2 = "kelb"
Output: true
Explanation: The two strings are already equal, so no string swap operation is required.
```

**Constraints:** 

- <code>1 <= s1.length, s2.length <= 100</code>
- <code>s1.length == s2.length</code>
- <code>s1</code> and <code>s2</code> consist of only lowercase English letters.

````python
from typing import List

def areAlmostEqual(s1: str, s2: str) -> bool:
    count = 0

    if (sorted(s1) == sorted(s2)):
        for i in range(len(s1)):
            if (s1[i] != s2[i]):
                count += 1
    else:
        return False

    return count == 0 or count == 2
````

````python
areAlmostEqual("bank", "kanb")
````
