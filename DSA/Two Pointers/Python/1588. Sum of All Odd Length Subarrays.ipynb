# [1588. Sum of All Odd Length Subarrays](https://leetcode.com/problems/sum-of-all-odd-length-subarrays/description/)

Two Approaches:
- Brute Force
- Math Formula for finding No. of Left and Right Subsets for the element

Given an array of positive integers <code>arr</code>, return the sum of all possible **odd-length subarrays**  of <code>arr</code>.

A **subarray**  is a contiguous subsequence of the array.

**Example 1:** 

```
Input: arr = [1,4,2,5,3]
Output: 58
Explanation: The odd-length subarrays of arr and their sums are:
[1] = 1
[4] = 4
[2] = 2
[5] = 5
[3] = 3
[1,4,2] = 7
[4,2,5] = 11
[2,5,3] = 10
[1,4,2,5,3] = 15
If we add all these together we get 1 + 4 + 2 + 5 + 3 + 7 + 11 + 10 + 15 = 58```

**Example 2:** 

```
Input: arr = [1,2]
Output: 3
<b>Explanation: </b>There are only 2 subarrays of odd length, [1] and [2]. Their sum is 3.```

**Example 3:** 

```
Input: arr = [10,11,12]
Output: 66
```

**Constraints:** 

- <code>1 <= arr.length <= 100</code>
- <code>1 <= arr[i] <= 1000</code>

**Follow up:** 

Could you solve this problem in O(n) time complexity?

````python
def sumOddLengthSubarrays(arr):
    n = len(arr)
    total = 0

    for i in range(n):
        j = i+1
        while(j <= n):
            total += sum(arr[i:j])
            j += 2
        
    return total
````

````python
# Main
print(sumOddLengthSubarrays([1,4,2,5,3]))  # 58
````

````python
sumOddLengthSubarrays([10,11,12])  # Example usage
````

Total subarrays containing 
```python 
arr[i] = (i + 1) × (n − i)
```

Odd-length subarrays
```python
arr[i] = ((i + 1) × (n − i) + 1) // 2
```

## Setup

Let:

* `n = 5`
* `arr = [1, 4, 2, 5, 3]`
* Pick index **`i = 2`** (value `2`)

```
Index:  0   1   2   3   4
Array: [1,  4,  2,  5,  3]
                ↑
               i = 2
```

## Step 1: Count left and right choices

### Left choices (including index i)

| Left boundary index | Elements included |
| ------------------- | ----------------- |
| 2                   | [2]               |
| 1                   | [4, 2]            |
| 0                   | [1, 4, 2]         |

Total left choices = **3 = i + 1**

---

### Right choices (including index i)

| Right boundary index | Elements included |
| -------------------- | ----------------- |
| 2                    | [2]               |
| 3                    | [2, 5]            |
| 4                    | [2, 5, 3]         |

Total right choices = **3 = n − i**

---

### Total subarrays containing `arr[i]`

Each left choice pairs with each right choice:

| Left | Right | Subarray    | Length |
| ---- | ----- | ----------- | ------ |
| 2    | 2     | [2]         | 1      |
| 2    | 3     | [2,5]       | 2      |
| 2    | 4     | [2,5,3]     | 3      |
| 1    | 2     | [4,2]       | 2      |
| 1    | 3     | [4,2,5]     | 3      |
| 1    | 4     | [4,2,5,3]   | 4      |
| 0    | 2     | [1,4,2]     | 3      |
| 0    | 3     | [1,4,2,5]   | 4      |
| 0    | 4     | [1,4,2,5,3] | 5      |

Total subarrays = **3 × 3 = 9**

---

## Step 2: Why “half (rounded up)” are odd-length

### Count odd vs even lengths

| Length | Odd / Even |
| -----: | ---------- |
|      1 | Odd        |
|      2 | Even       |
|      3 | Odd        |
|      2 | Even       |
|      3 | Odd        |
|      4 | Even       |
|      3 | Odd        |
|      4 | Even       |
|      5 | Odd        |

* Odd-length subarrays = **5**
* Even-length subarrays = **4**

Total = **9**

So:

```
odd_count = (9 + 1) // 2 = 5
```

## Why the formula always works

```
Total subarrays containing arr[i]
= (i + 1) × (n − i)

Odd-length subarrays
= ((i + 1) × (n − i) + 1) // 2
```

The `+1` handles cases where the total is odd (like 9), ensuring correct rounding.



````python
def sumOddLengthSubarrays2(arr):
    n = len(arr)
    total = 0

    for i in range(n):
        count = ((i+1) * (n-i) + 1) // 2  # Odd Length
        total += arr[i] * count

    return total
````
