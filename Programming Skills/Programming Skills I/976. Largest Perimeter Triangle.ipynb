Given an array of integers **nums** representing side lengths, return the **largest possible perimeter** of a triangle with **non-zero area**.

A triangle is valid **if and only if**:

> For any three sides a â‰¤ b â‰¤ c â†’ **a + b > c**

## ðŸ’¡ Key Insight

To maximize perimeter, you should try **largest sides first**.

So:

1. **Sort nums in descending order**
2. Iterate from largest to smallest
3. For each consecutive triple:

   * if `nums[i] < nums[i+1] + nums[i+2]` â†’ valid triangle
   * return their sum
4. If no triple works â†’ return 0

This is efficient because sorting is *O(n log n)*, scanning is *O(n)*.

# âœ… Python Solution (Clean + Optimal)

```python
class Solution:
    def largestPerimeter(self, nums):
        nums.sort(reverse=True)
        
        for i in range(len(nums) - 2):
            a, b, c = nums[i], nums[i+1], nums[i+2]
            if b + c > a:     # Triangle inequality
                return a + b + c
        
        return 0
```

````python
def largestPerimeter(nums) -> int:
    n = len(nums)
    nums.sort(reverse = True)  # Decreasing Order

    for i in range(n-2):
        window = nums[i: i+3]
        if (window[0] < window[1] + window[2]):
            return sum(window)

    return 0
````
