# [1491. Average Salary Excluding the Minimum and Maximum Salary](https://leetcode.com/problems/average-salary-excluding-the-minimum-and-maximum-salary/)

```python
def average(salary):
    minVal = min(salary)
    maxVal = max(salary)
    
    return (sum(salary) - minVal - maxVal) / (len(salary) - 2)
```

```python
lst = [100, 200, 400, 300]
print(average(lst))
```
