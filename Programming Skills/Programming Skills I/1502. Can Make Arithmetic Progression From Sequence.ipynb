# [1502. Can Make Arithmetic Progression From Sequence](https://leetcode.com/problems/can-make-arithmetic-progression-from-sequence/description/)

A sequence of numbers is called an **arithmetic progression**  if the difference between any two consecutive elements is the same.

Given an array of numbers <code>arr</code>, return <code>true</code> if the array can be rearranged to form an **arithmetic progression** . Otherwise, return <code>false</code>.

**Example 1:** 

```
Input: arr = [3,5,1]
Output: true
Explanation: We can reorder the elements as [1,3,5] or [5,3,1] with differences 2 and -2 respectively, between each consecutive elements.
```

**Example 2:** 

```
Input: arr = [1,2,4]
Output: false
Explanation: There is no way to reorder the elements to obtain an arithmetic progression.
```

**Constraints:** 

- <code>2 <= arr.length <= 1000</code>
- <code>-10^6 <= arr[i] <= 10^6</code>

````python
from typing import List

def canMakeArithmeticProgression(arr: List[int]) -> bool:
    flag = True
    
    arr.sort()

    for i in range(1,len(arr)-1):
        if (arr[i-1]-arr[i] != arr[i] - arr[i+1]):
            return False

    return flag
````

````python
canMakeArithmeticProgression([3,5,1])
````
