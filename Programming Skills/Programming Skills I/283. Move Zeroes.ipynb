# [283. Move Zeroes](https://leetcode.com/problems/move-zeroes/description/)

Given an integer array <code>nums</code>, move all <code>0</code>'s to the end of it while maintaining the relative order of the non-zero elements.

**Note**  that you must do this in-place without making a copy of the array.

**Example 1:** 

```
Input: nums = [0,1,0,3,12]
Output: [1,3,12,0,0]
```

**Example 2:** 

```
Input: nums = [0]
Output: [0]
```

**Constraints:** 

- <code>1 <= nums.length <= 10^4</code>
- <code>-2^31 <= nums[i] <= 2^31 - 1</code>

**Follow up:**  Could you minimize the total number of operations done?

````python
def moveZeroes(nums):
    """
    Do not return anything, modify nums in-place instead.
    """
    left = 0

    if 0 not in nums:
        return nums

    for right in range(len(nums)):
        if nums[right] != 0:
            nums[left], nums[right] = nums[right], nums[left]
            left += 1 
````

````python
# Main
if __name__ == "__main__":
    arr = [0, 1, 0, 3, 12]
    moveZeroes(arr)
    print(arr)  # Output should be [1, 3, 12, 0, 0]
````

````python
arr = [1,2,3]
moveZeroes(arr)
print(arr)  # Output should be [1, 2, 3]
````
